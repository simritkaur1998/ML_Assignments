<b>Data Fields Explanation</b><br />

The White Wine Quality dataset consists of 4898 entries of the wine samples. This dataset consists of a single CSV file, winequality-white.csv. The columns in the table are:
<pre>
    fixed acidity - presence of nonvolatile acids in the wine
    volatile acidity - the amount of acetic acid in wine
    citric acid - amount of citric acid present in the wine(usually added in small quantity)
    residual sugar - the amount of sugar present in the wine after fermentation
    chlorides - the amount of salt in the wine
    free sulfur dioxide - amount of the free form of SO2 present in the wine
    total sulfur dioxide - amount of free and bound forms of S02
    density - tells the density of the water
    pH - describes how acidic or basic a wine is
    sulphates - a wine additive which can contribute to sulfur dioxide gas (S02) levels
     alcohol - the percent alcohol(% by volume) content of the wine
</pre>
