```python
#1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dtrain=pd.read_csv(r'https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
dtest=pd.read_csv(r'https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
```

```python
dtrain.head()
```

```python
dtest.head()
```

```python
df=pd.concat([dtrain,dtest],sort=False)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#2
df = df.dropna(subset=['Survived'])
```

```python
df.shape
```

```python
#3
df['Has_Cabin']=~df['Cabin'].isnull()
```

```python
df['Has_Cabin']
```

```python
#4
#df['Title']=df['Name'].str.extract(', (\w*)\.')
import re
pattern = re.compile('(\w*[.])')
title_list = []
titles = ["Mr.", "Mrs.", "Master.", "Miss."]
for i in df['Name']:
    title = pattern.findall(i)
    title = ''.join(title) #converting list type to str
    if title in titles:
        title_list.append(title[0:-1])
    else:
        title_list.append('Special')
    
df['Title'] = title_list
```

```python
df['Title']
```

```python
#5
df=df.fillna(df.mean())
```

```python
df
```

```python
#6
df['Categorical_age'] = pd.qcut(df['Age'],4,labels=['children','young','middle','aged'])
df['Categorical_fare'] = pd.qcut(df['Fare'],4,labels =['very low','low','normal','high'])
df[['Categorical_age', 'Categorical_fare']]
```

```python
#7
df = df.drop(['Cabin','Name','PassengerId','Ticket','Age','Fare','SibSp','Parch'],1)
```

```python
#8
df=pd.get_dummies(df)
df
```

```python
#9
df.columns
```

```python
x=df.drop(["Survived"],axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
#10
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
#11
score = classifier.score(X_test, y_test)
print(score)
```
